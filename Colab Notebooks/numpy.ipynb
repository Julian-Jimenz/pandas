```python
import numpy as np

# Crear un array tridimensional aleatorio
array = np.random.randint(1, 10, size=(4, 3, 3))
print("Array tridimensional:")
print(array)

lista = []
i = 0

# Iterar sobre el primer eje de la matriz
while i < 4:
    # Seleccionar el subarray en la posición i
    subarray = array[i]

    # Filtrar los valores mayores o iguales a 5 y agregarlos a la lista
    elementos_filtrados = subarray[subarray >= 5]  # Esto selecciona solo los elementos >= 5
    lista.append(elementos_filtrados.tolist()) # Convertir a lista y añadir a la lista principal


    i += 1

print("\nLista de elementos mayores o iguales a 5:")
print(lista)
```

```python
#Instrucción: Crea dos arrays de 1D con valores enteros y realiza las operaciones de suma, resta, multiplicación, y división entre ellos.

import numpy as np

array_1 = np.array([1,2,3,4,5])
array_2 = np.array([6,7,8,9,10])

suma = array_1 + array_2
resta = array_2- array_1
multiplicacion = array_1 * array_2
division = array_2 / array_1

print('suma',suma)
print("resta",resta)
print("mutiplicacion",multiplicacion)
print("division",division)
```

```python
# Instrucción: Dado un array de datos, calcula la media, mediana, varianza, y desviación estándar.
import numpy as np

datos = np.array([23, 76, 35, 67, 89, 45, 68, 79, 35])
media = np.mean(datos)
mediana = np.median(datos)
varianza = np.var(datos)
desviasion = np.std(datos)

print('media',media)
print("mediana",mediana)
print("varianza",varianza)
print("desviacion",desviasion)

```

```python
# Instrucción: Crea dos matrices de 2x2 y realiza las operaciones de suma, resta, multiplicación (producto matricial)
# y cálculo de la inversa de una de ellas.

import numpy as np

matriz_1 = np.array([[4,7],[2,6]])
matriz_2 = np.array([[5,6],[7,8]])

suma = matriz_1 + matriz_2
resta = matriz_2 - matriz_1
multiplicacion = np.dot(matriz_1,matriz_2)
inversa = np.linalg.inv(matriz_1)
# Verificación de la inversa
verificacion = np.dot(matriz_1, inversa)

print('suma',suma)
print("resta",resta)
print("mutiplicacion",multiplicacion)
print("inversa",inversa)
```

```python
#Resolución de un Sistema de Ecuaciones Lineales

#Instrucción: Resuelve el sistema de ecuaciones lineales dado por Ax=b, donde A es una matriz 2x2 y b es un vector de 2 elementos.

#Ax=bAx = b


import numpy as np

A = np.array([[3, 2], [1, 4]])
b = np.array([1, 2])

x = np.linalg.solve(A, b)

print(x)
```
